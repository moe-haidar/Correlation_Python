```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matric for numeric feature')
plt.xlabel("Movie Feature")
plt.ylabel("Movie Feature")

plt.show()
```

```python
# Looking at text columns and convert them to numeric values using cat.codes functions
#(gives random codes to the values of the column keeping the same values with same codes)

df_numerized = df.copy()

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category') 
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head(100)
```

```python
df.head(100)
```

```python

correlation_matrix = df_numerized.corr(method='pearson')
plt.figure(figsize = (20,19))
sns.heatmap(correlation_matrix, annot = True, cmap='RdYlBu_r', fmt= '.4g')
plt.title('Correlation matric for numeric feature')
plt.xlabel("Movie Feature")
plt.ylabel("Movie Feature")

plt.show()
```
